```python
import requests

#Um API com fotos aleatórias de gatos!!!
response = requests.get('https://thatcopy.pw/catapi/rest/')

print("Veja uma foto de um gato:")
for data in response.json():
    print(data + ": " + str(response.json()[data]))
```
